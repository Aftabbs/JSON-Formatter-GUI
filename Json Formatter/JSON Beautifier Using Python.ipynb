```python
import json
import tkinter as tk
from tkinter import scrolledtext

def format_json(input_json):
    try:
        formatted_json = json.dumps(input_json, indent=4)
        return formatted_json
    except Exception as e:
        return f"Error formatting JSON: {str(e)}"

def display_gui(formatted_json):
    root = tk.Tk()
    root.title("JSON Formatter")

    text_area = scrolledtext.ScrolledText(root, wrap=tk.WORD, width=60, height=20)
    text_area.insert(tk.INSERT, formatted_json)
    text_area.pack(expand=True, fill="both")

    root.mainloop()

def main():
    unformatted_json = {
        "name": "John Doe",
        "age": 30,
        "city": "New York"
        # Add more key-value pairs as needed
    }

    formatted_json = format_json(unformatted_json)
    display_gui(formatted_json)

if __name__ == "__main__":
    main()
```

```python
import json
import tkinter as tk
from tkinter import scrolledtext

def format_json(input_json):
    try:
        formatted_json = json.dumps(input_json, indent=4)
        return formatted_json
    except Exception as e:
        return f"Error formatting JSON: {str(e)}"

def on_format_button_click():
    user_input = json_entry.get("1.0", tk.END)
    try:
        user_input_json = json.loads(user_input)
    except json.JSONDecodeError:
        formatted_output = "Invalid JSON input."
    else:
        formatted_output = format_json(user_input_json)
    output_text.config(state=tk.NORMAL)
    output_text.delete("1.0", tk.END)
    output_text.insert(tk.INSERT, formatted_output)
    output_text.config(state=tk.DISABLED)

root = tk.Tk()
root.title("JSON Beautifier")

json_entry_label = tk.Label(root, text="Enter JSON:")
json_entry_label.pack()
json_entry = scrolledtext.ScrolledText(root, wrap=tk.WORD, width=60, height=5)
json_entry.pack()

format_button = tk.Button(root, text="Format JSON", command=on_format_button_click)
format_button.pack()

output_text_label = tk.Label(root, text="Formatted JSON:")
output_text_label.pack()
output_text = scrolledtext.ScrolledText(root, wrap=tk.WORD, width=60, height=20, state=tk.DISABLED)
output_text.pack(expand=True, fill="both")

root.mainloop()
```
